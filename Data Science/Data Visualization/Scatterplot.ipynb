```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
#iki değişkenin arasındaki ilişkiyi incelemek için koralasyon kullanılır
sns.scatterplot(x = "total_bill",y = "tip",data = df);
```

```python
(sns.scatterplot(x = "total_bill",
                 y = "tip",
                 hue = "time",
                 style = "time",
                 data = df));
```

```python
(sns.scatterplot(x = "total_bill",
                 y = "tip",
                 hue = "day",
                 style = "time",
                 data = df));
```

```python
(sns.scatterplot(x = "total_bill",
                 y = "tip",
                 hue = "size",
                 size = "size",
                 data = df));
```
