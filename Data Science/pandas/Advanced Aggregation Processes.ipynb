```python
import pandas as pd
import numpy as np
df = pd.DataFrame({"gruplar":["A","B","C","A","B","C"],
                  "değer1":[10,20,34,45,54,54],
                   "değer2":[102,203,344,455,546,546]})
df
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min",np.median,max])
#sadece istediğimiz fonksiyonları çalıştırmamızı sağlar
#pandasta olmayan fonksiyonları "" arasında yazamazsın
```

```python
df.groupby("gruplar").aggregate({"değer1":min,"değer2":max})
#değer1 ve değer 2 için farklı fonksiyon yapar
```

```python
def fonk(df):
    df2 = df.groupby("gruplar").min()
    df2 = df2.iloc[0:2]
    return df2
fonk(df)
```

```python
def fonk2(df):
    df2 = df.groupby("gruplar").std()
    df2 = df2.iloc[0:2]
    return df2
fonk2(df)
```
