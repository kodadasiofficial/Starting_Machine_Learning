```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"isimler":["Adahan","Mehmet","Aysun","Ömer"],
                   "iş":["Mühendis","Asker","Bilim Adamı","Bilim Adamı"]})
df
```

```python
df2 = pd.DataFrame({"isimler":["Adahan","Mehmet","Aysun","Ömer"],
                   "doğum tarihi":[2004,1981,1982,2013]})
df2
```

```python
pd.merge(df,df2)#iki dataframede de aynı sütunlar varsa sadece birini alır
```

```python
pd.merge(df,df2,on = "isimler")#on hangi sütünun ortak alınacağını belirler
```

```python
df3 = pd.merge(df,df2)
```

```python
df3
```
