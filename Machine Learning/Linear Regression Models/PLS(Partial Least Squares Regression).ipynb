# MODEL KURMA 

```python
import pandas as pd
hitters = pd.read_csv("hitters.csv")
df = hitters.copy()
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.describe().T
```

```python
df.corr()
```

```python
import seaborn as sns
sns.pairplot(df,kind = "reg");
```

```python
dummies = pd.get_dummies(df[["League","Division","NewLeague"]])
dummies.head()
```

```python
dummies = dummies.drop(["League_A","Division_E","NewLeague_A"],axis = 1)
dummies.head()
```

```python
y = df["Salary"]
x_ = df.drop(["Salary","League","Division","NewLeague"],axis = 1).astype("float64")
x = pd.concat([x_,dummies],axis = 1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,
                                                       test_size = 0.25,
                                                       random_state = 42)
```

```python
from sklearn.cross_decomposition import PLSRegression,PLSSVD
```

```python
pls_model = PLSRegression().fit(x_train,y_train)
#n_components deikenine tuning yaparak bulduumuz bileen says girilir.
#Yani algoritmann veri setimizi ka deikene indirgeyeceini belirliyoruz.
```

```python
pls_model.coef_
#Burada modeldeki deikenlerin katsaylarn bulduk.
#Buradaki katsay saysnn bileen saysn deitirsek bile ayn kalmasnn 
#nedeni dardan gelen deerler deiken says kadar olaca iin onlarn hepsini tahmin edebilmek iin o kadar katsay olmas gerekliliidir.
#Dardan gelen deikenlerin katsaylar kullanlarak belli dorusal kombinasyonlar ile tahmin etmi oluyoruz.
#Yaplacak kombinasyonlara gre de bu katsaylar belirlenir.
```

# Tahmin 

```python
y_tahmin = pls_model.predict(x_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score
np.sqrt(mean_squared_error(y_train,y_tahmin))
```

```python
r2_score(y_train,y_tahmin)
```

```python
y_tahmin_test = pls_model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_tahmin_test))
```

```python
r2_score(y_test,y_tahmin_test)
```

# Model Tuning

```python
from sklearn import model_selection
cv = model_selection.KFold(n_splits = 10,shuffle = True,random_state = 1)
```

```python
rmse = []
for i in np.arange(1,x_train.shape[1]+1):
    pls = PLSRegression(n_components = i)
    score = np.sqrt(-1*model_selection.cross_val_score(pls,x_train,y_train,
                                                       cv = cv,
                                                      scoring = "neg_mean_squared_error").mean())
    rmse.append(score)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(np.arange(1,x_train.shape[1]+1),np.array(rmse),"-v",c = "r")
plt.xlabel("Bileen Says")
plt.ylabel("RMSE Deerleri")
plt.title("Salary Tahmin Tuning");
```

```python
#cross validation yntemi kullanarak bulduumuz hata kareler ortalamasnn 
#karekk deerlerine gre en optimum hata orann aldmz deerin 2 olduunu
#gryoruz yani bileen saysn 2 olarak verip modelimizi eittiimizde
#en optimum hata oranna ulaabiliyoruz.
```

```python
pls_model = PLSRegression(n_components = 2).fit(x_train,y_train)
```

```python
y_tahmin_test = pls_model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_tahmin_test))
```

```python
r2_score(y_test,y_tahmin_test)
```
