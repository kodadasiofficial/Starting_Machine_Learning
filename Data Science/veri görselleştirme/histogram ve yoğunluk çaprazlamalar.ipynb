```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
elmas = sns.load_dataset("diamonds")
df = elmas.copy()
```

```python
kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = kategoriler,ordered = True))
```

```python
sns.kdeplot(df.price,shade = True);
```

```python
(sns.FacetGrid(df,
              hue = "cut",
              height = 5,
              xlim = (0,10000),#odaklanan aralk
            ).map(sns.kdeplot,"price",shade = True)
.add_legend());
#facetgrid fonksiyonunu kdeplot ile birletirdik 
#ve younluk fonksiyonuna yeni bir boyut ekledik
```

```python
sns.catplot(x = "cut",y = "price",hue = "color",kind = "point", data = df);
```
