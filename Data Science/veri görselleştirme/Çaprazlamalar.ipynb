```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
elmaslar = sns.load_dataset("diamonds")
df = elmaslar.copy()
kategori = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = kategori,ordered = True))
df.head()
```

```python
sns.catplot(x = "cut",y = "price",data = df);
```

```python
df.cut.head(1)
```

```python
sns.barplot(x = "cut" , y = "price" , hue = "color",data = df);
#3.boyutu eklediğinde sutün grafiği ortalamalara göre çizilir
```

```python
#yukarıdaki grafiği doğrulamak için
df.groupby(["cut","color"])["price"].aggregate(["mean","std"])
```
